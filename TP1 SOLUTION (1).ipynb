```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Chargement des données
# Note: Assurez-vous que le fichier est dans le même dossier
df = pd.read_csv('AmesHousing.csv')

# 2. Imputation des valeurs manquantes (Numériques)
num_VM = ['Lot Frontage', 'Mas Vnr Area', 'BsmtFin SF 1', 'BsmtFin SF 2', 'Bsmt Unf SF', 
          'Total Bsmt SF', 'Bsmt Full Bath', 'Bsmt Half Bath', 'Garage Yr Blt', 'Garage Cars', 'Garage Area']

for n_col in num_VM:
    if n_col in df.columns:
        df[n_col] = df[n_col].fillna(df[n_col].mean())

# 3. Imputation des valeurs manquantes (Catégorielles)
colonnes_avec_VM = df.columns[df.isna().sum() > 0]
nom_VM = [x for x in colonnes_avec_VM if x not in num_VM]

for nom_col in nom_VM:
    if not df[nom_col].mode().empty:
        df[nom_col] = df[nom_col].fillna(df[nom_col].mode()[0])

# 4. Encodage et Sélection
df['MS SubClass'] = df['MS SubClass'].astype(str)
selection = ['MS SubClass', 'MS Zoning', 'Lot Frontage', 'Lot Area', 'Neighborhood', 
             'Overall Qual', 'Overall Cond', 'Year Built', 'Total Bsmt SF', '1st Flr SF', 
             '2nd Flr SF', 'Gr Liv Area', 'Full Bath', 'Half Bath', 'Bedroom AbvGr', 
             'Kitchen AbvGr', 'TotRms AbvGrd', 'Garage Area', 'Pool Area', 'SalePrice']

df = df[selection]
df = pd.get_dummies(df)

# 5. Partitionnement et Standardisation
train = df.sample(frac=0.8, random_state=9)
test = df.drop(train.index)

train_cible = train.pop('SalePrice')
test_cible = test.pop('SalePrice')

variables_pred = train.columns
for col in variables_pred:
    moy, std = train[col].mean(), train[col].std()
    if std < 1e-20: std = 1e-20
    train[col] = (train[col] - moy) / std
    test[col] = (test[col] - moy) / std

print("Prétraitement terminé avec succès !")
df.head()
```
